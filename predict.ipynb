```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import pickle
from tensorflow.keras.models import load_model
from sklearn.compose import ColumnTransformer
```

```python
model = load_model('model.h5')

with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('one_hot_encoder_geography.pkl', 'rb') as file:
    one_hot_encoder_georgaphy = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
# Example of input Data

input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male', 
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}

input_df = pd.DataFrame([input_data])
```

```python
geo_encoded = one_hot_encoder_georgaphy.transform(
    input_df[['Geography']]
)

geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=one_hot_encoder_georgaphy.get_feature_names_out(['Geography'])
)

geo_encoded_df
```

```python
# Encode my Categorical Variable

input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df
```

```python
# Concatenation of One Hot Encoded Data

input_df = pd.concat([input_df.drop("Geography", axis=1), geo_encoded_df], axis=1)
```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
# Prediction
prediction = model.predict(input_scaled)
```

```python
prediction_proba = prediction[0][0]
```

```python
if prediction_proba > 0.5:
    print('The Customer is Likely to Churn')
else:
    print('The Customer is not Likely to Churn')
```
